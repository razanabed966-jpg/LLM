````python
def get_completion(prompt, model="gpt-4.1-mini"):
    messages = [
              {"role": "user",
              "content": prompt}
               ]
    Response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, # model output randomness
    )
    return Response.choices[0].message.content
````
