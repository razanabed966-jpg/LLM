```python
import transformers
import torch
```

```python
model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)
```

```python
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]

outputs = pipeline(
    messages,
    max_new_tokens=256,
)
print(outputs[0]["generated_text"][-1])
```

```python
def run_chat(pipeline, user_input, system_message="You are a helpful assistant!", max_new_tokens=256):

    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": user_input}
    ]

    outputs = pipeline(
        messages,
        max_new_tokens=max_new_tokens
    )

    return outputs[0]["generated_text"][-1]["content"]
```

```python
response = run_chat(pipeline, "What is the capital of France?")
print(response)
```

```python
response = run_chat(pipeline, "The sum of the numbers 1 and 25 equal to")
print(response)
```

```python
response = run_chat(pipeline, "ما هي عاصمة فرنسا؟")
print(response)
```
